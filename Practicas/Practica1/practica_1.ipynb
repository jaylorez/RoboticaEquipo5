### URDF (Unified Robot Description Format):
El URDF es un formato de archivo XML que se utiliza para describir la estructura de un robot, comunmente usado en ROS. Permite definir la composici√≥n f√≠sica de un robot, incluyendo sus eslabones y juntas. Adem√°s, permite especificar caracter√≠sticas como el peso, la geometr√≠a, y la disposici√≥n espacial de los elementos del robot.

Los elementos primitivos son formas geom√©tricas b√°sicas que se utilizan para representar las partes de un robot en URDF. Algunos de los m√°s comunes son:

- **Caja (box)**: Un paralelep√≠pedo definido por su ancho, alto y profundidad.
- **Cilindro (cylinder)**: Definido por su radio y longitud.
- **Esfera (sphere)**: Definida por su radio.
- **Malla (mesh)**: Permite importar geometr√≠as m√°s complejas en formatos como STL, Collada, obj, entre otros.

En el formato URDF, existen varios tipos de juntas que pueden definir el movimiento relativo entre las partes de un robot:

- **Junta fija**: No permite ning√∫n movimiento entre los elementos conectados.
- **Junta rotacional**: Permite un movimiento de rotaci√≥n alrededor de un eje.
- **Junta prism√°tica**: Permite un movimiento lineal a lo largo de un eje.
- **Junta continua**: Es una variaci√≥n de la junta rotacional que permite una rotaci√≥n sin l√≠mites.

- ¬øQu√© es un archivo URDF?

El formato URDF (Unified Robot Description Format) es una especificaci√≥n basada en XML que permite describir la estructura y caracter√≠sticas de un robot. A trav√©s de este formato, es posible definir su aspecto visual, su cinem√°tica y din√°mica, as√≠ como su modelo de colisi√≥n. Gracias a estas descripciones, se pueden realizar simulaciones de robots en entornos como GAZEBO o RVIZ2, lo que facilita el dise√±o y prueba de sus funcionalidades antes de su implementaci√≥n f√≠sica.
Para representar un robot en URDF, se descomponen sus partes en enlaces (links) y articulaciones (joints). Cada una de las partes r√≠gidas que componen el robot se denomina eslab√≥n o link, mientras que la forma en que estos enlaces se conectan entre s√≠ constituye una articulaci√≥n o joint. Estas articulaciones determinan el tipo de movimiento que puede realizar el robot y se clasifican en diferentes tipos seg√∫n sus restricciones y grados de libertad.
Entre los tipos de articulaciones permitidas en URDF se encuentran las revolute, que permiten el giro sobre un eje con un rango limitado definido por valores superior e inferior, y las continuous, que tambi√©n permiten el giro pero sin restricciones de rango. Asimismo, existen las articulaciones prismatic, que permiten el deslizamiento a lo largo de un eje con un rango de movimiento limitado, y las fixed, que no permiten movimiento alguno pero sirven para establecer puntos de referencia en sensores o c√°maras. Tambi√©n se encuentran las floating, que permiten movimiento en los seis grados de libertad, y las planar, que restringen el movimiento a un plano perpendicular a un eje determinado.
Dentro de la estructura del robot, los eslabones unidos por una articulaci√≥n se organizan de manera jer√°rquica. El enlace m√°s cercano a la base o chasis del robot se denomina padre, mientras que el otro eslab√≥n conectado recibe el nombre de hijo. Un ejemplo de esta organizaci√≥n se puede observar en el robot manipulador IRB120 de ABB, el cual est√° compuesto por seis eslabones y seis articulaciones de giro, lo que le permite realizar movimientos precisos y controlados en aplicaciones industriales.
